```python
onlyfiles = [f for f in os.listdir('.') if isfile(join(".", f))]
onlyfiles
```

```python
author_virus = ['Schneider - HCoV-229E',
 'Kulsuptrakul - HAV',
 'Zhang - WNV',
 'Zhang - WNV',
 'Schneider - HCoV-NL63',
 'Marceau - DENV 216681 strain',
 'Marceau - DENV 216681 strain',
 'Wang - SARS-CoV-2',
 'Flint - EBOV',
 'Wang - HCoV-229E',
 'Marceau - DENV 216681 strain',
 'Zhang - WNV',
 'Biering - SARS CoV2',
 'Wang - HCoV-OC43',
 'Scheider - SARS CoV2',
 'Biering - SARS CoV2',
 'Biering - SARS CoV2',
 'Diep - RV C15',
 'Diep - EV D68',
 'Scheider - SARS CoV2',
 'Marceau - DENV 216681 strain',
 'Kulsuptrakul - HAV',
 'Flint - EBOV'
]

df_list = list()
for f in onlyfiles:
    if f != ".DS_Store":
        df = pd.read_csv(f)
        df_list.append(df)

#Compiles two lists into a dictionary, such that "author name - virus" is the key and the file is the value.
dict1 = dict(zip(author_virus, df_list))

result = pd.concat(dict1)
print(result)

result.to_csv('/Users/elianna.kondylis/Desktop/out3.csv')

# The Jupyter file does not show the entire metadata collumn, however, the .csv file includes it.
```
